```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display # to display images
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/admin/OneDrive - Dong Nai Technology University (DNTU)/TGMT/ThuVien/images5.jpg'
```

```python
# Đọc ảnh 
img = load_image(my_path)
```

```python
display(img)
```

```python
# Kích thước ban đầu:
print(img.size)
```

```python
# Thay đổi kích thước
new_size = (600,350)
new_image = img.resize(new_size)
display(new_image)
# Task: lưu ảnh
```

```python
# Thay đổi kích thước
thumbnail_size = (50,50)
img.thumbnail(thumbnail_size)
display(img)
# Task: lưu ảnh
```

```python
# Định vị khu vực cắt
region = (100,75,175, 150)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (150,150)
img.paste(cropped_image, paste_position)
display(img)
```

```python
# Xoay ảnh
rotated_image = img.rotate(90)
display(rotated_image)
```

```python
rotated_image = rotated_image.rotate(90)
display(rotated_image)
```

```python
# Transpose ảnh
transposed_image = img.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```

```python
# Transpose ảnh
transposed_image = transposed_image.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
