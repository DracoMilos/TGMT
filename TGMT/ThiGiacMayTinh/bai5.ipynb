```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#Duong dan anh 
my_path = 'C:/Users/admin/OneDrive - Dong Nai Technology University (DNTU)/TGMT/ThuVien/images1.jpg'
#Doc anh 
im = Image.open(my_path)
plt.imshow(im)
```

```python
#chuyensangsam
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#Tao mot hinh ve moi 
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title('Biểu đồ Histogram của ảnh ')
plt.xlabel('giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
#Lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
